```python
import matplotlib.pyplot as plt
import matplotlib.image as mping
import numpy as np
import random
import cv2




canvas = np.ones((500,500,3), dtype=np.uint8)
x = np.linspace(-2, 2, 500)  # go from -2 to 2 for a smooth tanh
tanh_gradient = np.tanh(x)   # values between -1 and 1
# Normalize tanh to 0-255 range
normalized = ((tanh_gradient + 1) / 2) * 255   # now between 0 and 255
for i in range(500):
    # set each column's color based on tanh gradient
    canvas[:, i, 0] = normalized[i]   # R channel has tanh gradient
    canvas[:, i, 1] = 200             # G channel is constant
    canvas[:, i, 2] = 130             # B channel is constant

plt.imshow(canvas)
plt.axis('off')
plt.show()
```
