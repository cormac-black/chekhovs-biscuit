```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mping

img_path = '/content/Images from Colab/headshot.jpg'

image = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_h, image_w = image.shape[:2]
print(image_h, image_w)


radians_sine = np.linspace(np.pi/2,2*np.pi+np.pi/2, image_h)
sine_wave = np.sin(radians_sine)
normalized_sin = ((sine_wave + 1)/2)* 255

radians_cos = np.linspace(np.pi/2,2*np.pi+np.pi/2, image_w)
cos_wave = np.cos(radians_cos)
normalized_cos = ((cos_wave + 1)/2)* 255

slope_tan = np.linspace(-2,2,image_h)
tanh_gradient = np.tanh(slope_tan)
normalized_tanh = ((tanh_gradient + 1)/2)* 255



for i in range(image_h):
  for j in range(image_w):

    image[i, :100, 0] = normalized_sin[i]
    image[:, :100, 1] = 200
    # image[:, j, 1] = normalized_cos[j]
    image[:, :100, 2] = 200
    # image[i, :, 2] = normalized_tanh[i]

plt.imshow(image)
plt.axis('off')
print('Image 1\n')
plt.show()
```
