```python
import numpy as np
import random


def make_array(h,w):
  h_enter = h
  w_enter = w
  g = np.zeros((h_enter,w_enter), dtype=int)

  hgt, wdt = g.shape

  for i in range(hgt):
      for j in range(wdt):
          a = random.randint(0,9)
          g[i,j] = a

  return g

a = make_array(5,5)
print(f'Matrix A:\n {a}\n\n')
b = make_array(5,5)
print(f'Matrix B: \n {b}\n')

try:
  y,x = np.where(a == 2)
  print(f'Matrix AB:\n {y,x}')
except:
  print("sorry no 2")
```
